```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_excel("回归数据.xlsx")
data=data[["证券名称","总资产(元)"]]
for i in range(len(data)):
    if data["总资产(元)"][i]=='——':
        data["总资产(元)"][i]=data["总资产(元)"][i-1]
```

```python
names=[]
for i in data.drop_duplicates(subset='证券名称', keep='first', inplace=False)["证券名称"]:
    names.append(i)
```

```python
meandata={}
for i in names:
    meandata[i]=data[data["证券名称"]==i]["总资产(元)"].mean()
```

```python
weightdata={}
for i in meandata:
    weightdata[i]=meandata[i]/sum(meandata.values())
```

```python
returndata=pd.read_excel("收益率.xlsx")
```

```python
renames=names
```

```python
renames.insert(0,"rm")
```

```python
renames.insert(0,"time")
```

```python
returndata_rename=returndata
```

```python
returndata_rename.columns=renames
```

```python
rm=[]
for j in range(len(returndata_rename)):
    a={}
    for i in weightdata:
        a[i]=weightdata[i]*returndata_rename[i][j]
    rm.append(sum(a.values()))
    
```

```python
returndata_rename["rm"]=rm
```
